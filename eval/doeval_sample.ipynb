```python
def find_mkey(mod_dir):
    for f in os.listdir(mod_dir):
        if f.endswith("final.model.pkl"):
            return f.split(".")[1]
    return None
```

```python
scmap_str = json.loads(open("./DataPrep/scmap.json", "r").read())
scmap = {int(k): v for k, v in scmap_str.items()}
scmap[0xffff] = "NOSC"
```

```python
blurs = [0, 4, 12]
n_feat = 64
do_log = 0
step = 400000
typs = ['pr', 'pw']
#typ = 'pr'

n_feats = [128, 512, 16,  256,  32,  384,  48,  64,  8,  96]
#n_feats = [512, 256, 128, 64, 32]
#key = 'ssh_0_xxx'
#key2 = 'ssh_0'

#data_path = "./data/memaccess_dumps/%s_%s.dump" % (key, typ)
#anot_path = "./data/memaccess_dumps/%s_%s.annotations" % (key, typ)
#info_path = "./data/recordings/%s.info" % key2
#mae = DataPrep.MaEval.MaEval(data_path, anot_path, info_path, scmap)
```

```python
key = "sample"
nosys_id = "3"
for typ in typs:
    print("-" * 80)
    print("TYP: %s" % typ)
    for blur in blurs:
        if typ == "pr" and blur == 4:
            continue
        hdkeys = []
        sample_fn = os.path.join("new_hdf", "%s_eval1_nosys%s_ei_b_%d_%d.dd.drop.nosys.hdf5" % (typ, nosys_id, blur, 512))
        with h5py.File(sample_fn, mode='r') as fd:
            for hdkey in fd.keys():
                if hdkey == "INFO":
                    continue
                if len(hdkey) <= 4:
                    continue
                hdkeys.append(hdkey)
        for n_feat in n_feats:
            model_dir = "fmod/%s/%d/%d" % (typ, blur, n_feat)
            if not os.path.exists(model_dir):
                continue
            mkey_hash = find_mkey(model_dir)
            if mkey_hash is None:
                continue
            mkey = "%s_%d_%d_Random_Forrest.%s" % (typ, n_feat, blur, mkey_hash)
            eval_fn = "%s_%s.eval%s_sample" % (key, mkey, nosys_id)
            eval_loaded = False
            if os.path.exists(eval_fn):
                try:
                    eval_data = pickle.load(open(eval_fn, "rb"))
                    y_pred = eval_data['y_pred']
                    y_base = eval_data['y_base']
                    sc_count_pred = eval_data['sc_count_pred']
                    sc_count_base = eval_data['sc_count_base']
                    all_sc = eval_data['all_sc']
                    key2sc = eval_data['key2sc']
                    sc2key = eval_data['sc2key']
                    cf_matrix = eval_data['cf_matrix']
                    eval_loaded = True
                except:
                    pass
                #continue
            if not eval_loaded:
                model_base_path = os.path.join(model_dir, mkey)
                print(model_base_path)
                print("Params: nfeat %d, blur %d" % (n_feat, blur))
                ncnr_path = "%s.NR2CNR.json" % model_base_path
                sc2key_str = json.loads(open(ncnr_path, "r").read())
                sc2key = {int(v): int(k) for v, k in sc2key_str.items()}
                key2sc = {int(v): int(k) for k, v in sc2key.items()}
                model_conf_path = "%s.py_caret_setup_config.pkl" % model_base_path
                model_path = "%s.final.model" % model_base_path
                print("Loading Config %s" % model_conf_path)
                pc.classification.load_config(model_conf_path)
                print("Loading Model %s" % model_path)
                pmdl = pc.classification.load_model(model_path)
                y_base = []
                y_pred = []
                all_sc = set()
                sc_count_pred = {}
                sc_count_base = {}
                for hdk_idx, hdk in enumerate(hdkeys):
                    sys.stdout.write("%4d %s\r" % (hdk_idx, hdk))
                    if '65535' in hdk:
                        tmpfn = "%s_%s_%d__%s.tmpsample%s" % (key, mkey, n_feat, hdk, nosys_id)
                    else:
                        tmpfn = "%s_%s_%d__%s.tmpsample2" % (key, mkey, n_feat, hdk)
                        if not os.path.exists(tmpfn):
                            tmpfn = "%s_%s_%d__%s.tmpsample" % (key, mkey, n_feat, hdk)
                    if os.path.exists(tmpfn):
                        tmp_data = pickle.load(open(tmpfn, "rb"))
                        y_pred_p = tmp_data['y_pred_p']
                        y_base_p = tmp_data['y_base_p']
                    else:
                        cols = ['KEY']
                        feature_cols = ['MAD_%d' % t for t in range(1, n_feat)]
                        cols.extend(feature_cols)
                        df = pd.read_hdf(sample_fn, key=hdk)[cols]
                        y_base_p = [sc2key[sc] for sc in df['KEY'].values]
                        y_pred_p = pmdl.predict(X=df[feature_cols])
                        pickle.dump({
                            'y_pred_p': y_pred_p,
                            'y_base_p': y_base_p,
                        }, open(tmpfn, "wb"))
                        
                    y_base.extend(y_base_p)
                    y_pred.extend(y_pred_p)
                    
                    (unique, counts) = np.unique(y_pred_p, return_counts=True)
                    for i in range(len(unique)):
                        sc = key2sc[unique[i]]
                        if sc not in sc_count_pred.keys():
                            sc_count_pred[sc] = 0
                        sc_count_pred[sc] += counts[i]
                        all_sc.add(sc)
                        
                    (unique, counts) = np.unique(y_base_p, return_counts=True)
                    for i in range(len(unique)):
                        sc = key2sc[unique[i]]
                        if sc not in sc_count_base.keys():
                            sc_count_base[sc] = 0
                        sc_count_base[sc] += counts[i]
                        all_sc.add(sc)
                        
                cf_matrix = confusion_matrix(y_base, y_pred, normalize='true')
                eval_data = {
                    "sc2key": sc2key,
                    "key2sc": key2sc,
                    "blur": blur,
                    "n_feat": n_feat,
                    "typ": typ,
                    "mkey": mkey,
                    "key": key,
                    "y_base": y_base,
                    "y_pred": y_pred,
                    "cf_matrix": cf_matrix,
                    "sc_count_pred": sc_count_pred,
                    "sc_count_base": sc_count_base,
                    "all_sc": list(all_sc),
                }
                pickle.dump(eval_data, open(eval_fn, "wb"))
            target_names = []
            sc_sorted = list(all_sc)
            sc_sorted.sort()
            keys_sorted = []
            for sc in sc_sorted:
                target_names.append(scmap[sc] + " (%d)" % sc)
                sc_excluded = False
                #if typ == "pw":
                #    sc_excluded = sc in ex_pw_id
                #if typ == "pr":
                #    sc_excluded = sc in ex_pr_id
                #if sc != 0xffff and not sc_excluded:
                #    keys_sorted.append(sc2key[sc])
                keys_sorted.append(sc2key[sc])
            keys_sorted_nosc = keys_sorted[:]
            keys_sorted_nosc.append(sc2key[0xffff])
            print("*" * 80)
            print("BLR: %d, NFEAT: %d, %s, %s" % (blur, n_feat, typ, key))
            sns.heatmap(cf_matrix, annot=False, cmap='Blues')
            plt.show()
            print(classification_report(y_base, y_pred, target_names=target_names, labels=keys_sorted))
            eval_data['class_rep'] = classification_report(y_base, y_pred, target_names=target_names, labels=keys_sorted, output_dict=True)
            pickle.dump(eval_data, open(eval_fn, "wb"))
```
